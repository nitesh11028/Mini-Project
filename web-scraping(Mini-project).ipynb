```python
from bs4 import BeautifulSoup
import requests
import csv
import pandas as pd
```

```python
url = 'https://www.empireonline.com/movies/features/best-movies-2/.'
```

```python
result = requests.get(url)
```

```python
# Explicitly set the encoding to 'utf-8'
result.encoding = 'utf-8'
```

```python
page_content = result.content
```

```python
soup = BeautifulSoup(page_content,'html.parser')
soup.contents
```

```python
result.close()
```

```python
movies = soup.find_all('div', class_='listicle_listicle__item__CJna4')
```

```python
# Extract movie names and ranks
movie_list = []
for movie in movies:
    rank = movie.find('span', class_='listicleItem_listicle-item__count__ofTIR').text.split()[0]
    name = movie.find('h3', class_='listicleItem_listicle-item__title__BfenH').text.strip().split(' ', 1)[1]
    movie_list.append((int(rank), name))
```

```python
# Sort the movies by rank in ascending order
movie_list.sort()
```

```python
movie_list
```

```python
# Save the data to a CSV file
# Save the data to a CSV file
with open('movies.csv', 'w', newline='',encoding='utf-8') as csvfile:
    csvwriter = csv.writer(csvfile)
    csvwriter.writerow(['Rank', 'Movie Name'])
    csvwriter.writerows(movie_list)
print("Movies have been successfully scraped and saved to movies.csv")
```
